### Task 9: What is the equivalent of `enumerate` for numpy arrays? Illustrate the syntax by iterating through a 2D array that you create.

```python
array = np.random.randint(1, 10, size=(2, 10))
#for index, value in np.ndenumerate(array):
    #print(index, value)
for index in range(len(array)):
    print(array[index])
print('\n')
for index, value in np.ndenumerate(array):
    print(index, value)
```

### Task 10: How to find common values between two arrays?

```python
array1 = np.random.randint(1, 10, size=(1, 10))
array2 = np.random.randint(1, 10, size=(1, 10))
print(array1, '\n', array2)
np.intersect1d(array1, array2)
```
