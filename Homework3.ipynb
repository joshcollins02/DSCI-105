### Task 12: Generate a new column named "Size" to classify countries based on their 2022 population. If a country had a population of 50 million or less, label it as "Small" otherwise label it as "Big".
<i> Hint: You can use .cut() function and float("inf") as the upper bound. </i>

```python
bins = [0,50000000,float('inf')]
labels = ['Small','Big']
population['Size'] = pd.cut(population['2022 Population'], bins=bins, labels=labels)
print(population['Size'])

#population['Size'] = pd.cut(population['2022 Population'], bins=[0,50000000,float('inf'), labels=['Small','Big'])
```
